```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("Salary_dataset.csv")
df.head()
```

```python
df.columns
```

```python
X=df[["Years Of Experience"]]
Y=df[["Salary"]]

print(X.head())
print(Y.head())
```

```python
model=LinearRegression() # model constructor same as linear class initialization
```

```python
model.fit(X,Y) # x=input, y=output
```

```python
# y = mx + c // 9449(2) + 24k = y
# output = slope(input) + constant 
print(model.coef_[0]) # experience 1 = 9k
print(model.intercept_) # rapido = 50Rs
```

```python
result = model.predict([[5]]) # 24848 = 9k+9k = 
print(result[0])
#119347 -10
```

```python
v = model.predict(X)
score = r2_score(v,Y)
print(score)
```

```python
# Label Encoding Implementation cannot be done because it needs text, but here there is no text.
#Random State : 80/20, It will take 80% training data from the dataset randomly and another 20% testing data randomly
```

```python
# 24/4/26

plt.scatter(X,Y,color = "blue")
plt.plot(X,model.predict(X),color = "red",marker=".")
plt.xlabel("exp")
plt.ylabel("Salary")
plt.title("expe vs salary")
plt.show()
```

```python
# data split : we need 4 parts namely x train, x test, y train, y test
# where x means input and y means output and method name is train test split

X_train,X_test, Y_train,Y_test = train_test_split(
    X,Y,test_size = 0.20, random_state = 42
)
model1 = LinearRegression()
```

```python
model1.fit(X_train,Y_train)
```

```python
print(model1.coef_[0])
print(model1.intercept_)
```

```python
result = model1.predict([[5]])
print(result[0])
```

```python
v = model1.predict(X_test)
score = r2_score(v,Y_test)
print(score)
```

```python
plt.scatter(X_train,Y_train,color = "blue")
plt.plot(X_train,model1.predict(X_train),color = "red")
plt.xlabel("exp")
plt.ylabel("Salary")
plt.title("expe vs salary")
plt.show()
```
